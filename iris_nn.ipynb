```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/Datasets/iris.csv')
data
```

```python
data=data[data['Species']!='Iris-virginica']
data
```

```python
data['Species']=data['Species'].astype('category')
data['Species']=data['Species'].cat.codes
data
```

```python
x=data.drop(columns=['Species'])
y=data['Species']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#initialising
alpha=0.01
iterations=1000000
w1=np.zeros([4,2])
w2=np.zeros([3,2])
```

```python
hidden_layer=np.zeros([3,1])
hidden_layer[0]=1   #setting bias
output_layer=np.zeros([2,1])
target=np.array([[0],[1]])
delta_hidden=np.zeros([3,1])
```

```python
def sigmoid(a):
    return 1/(1 + np.exp(-a))
```

```python
for k in range(len(xtrain)):
    for i in range(1000):
        #feed forwarding
        x=xtrain.iloc[k][...,None]
        if ytrain.values[k]==0:
            target=np.array([[1],[0]])
        else:
            target=np.array([[0],[1]])
        hidden_layer[1:]=sigmoid(np.matmul(w1.T,x))
        output_layer=sigmoid(np.matmul(w2.T,hidden_layer))
        error=(output_layer-target)
        #back propagation
        delta_output=(output_layer-target)*output_layer*(1-output_layer)
        gradient=np.matmul(hidden_layer,delta_output.T)
        #weight updation
        w2=w2-alpha*gradient

        delta_hidden[1:]=np.matmul(w2[1:],delta_output)*hidden_layer[1:]*(1-hidden_layer[1:])
    #     delta_hidden[0]=np.matmul(w2[0],delta_output)
        gradient_hid=np.matmul(x,delta_hidden[1:].T)
        w1=w1-alpha*gradient_hid
```

```python
ypred=[]
for i in range(len(xtest)):
    x=xtest.iloc[i][...,None]
    hidden_layer[1:]=sigmoid(np.matmul(w1.T,x))
    output_layer=sigmoid(np.matmul(w2.T,hidden_layer))
    if output_layer[0][0]>output_layer[1][0]:
        ypred.append(0)
    else:
        ypred.append(1)
```
